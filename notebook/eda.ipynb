```python
# convert the object to numberic
cols = ["trestbps", "chol", "fbs", "restecg", "thalachh", 
        "exang", "slope", "ca", "thal"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
sns.histplot(data=df, x='slope', kde=True, hue='target')
```

slope is a categorical:
0 (Upsloping) Rare and less informative due to low frequency
1(Flat) More associated with target = 0 (healthy)
2(Downsloping) Strongly associated with target = 1 (heart disease)

```python
sns.histplot(data=df, x='ca', kde=True, hue='target')
```

The majority of observations are at ca =  0 (healthy) and requency decreases as ca increases (1 to 4) [right-skewed discrete distribution disease) 
increases higher ca values are strongly associated with presence of heart disease

```python
sns.histplot(data=df, x='thal', kde=True, hue='target')
```

Most values are in defect categories (2 & 3 / 6 & 7).
These defect categories are strongly associated with heart disease (target = 1).
The normal category is more linked with healthy patients (target = 0).
Overall, thal is a highly important predictor for heart disease.
